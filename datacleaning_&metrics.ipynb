## Create KPIs

```python
# Create KPIs
df_marketing['CTR'] = df_marketing['Clicks'] / df_marketing['Impressions']
df_marketing['CPC'] = df_marketing['Spend ($)'] / df_marketing['Clicks']
df_marketing['CPA'] = df_marketing['Spend ($)'] / df_marketing['Conversions']
df_marketing['ROI'] = (df_marketing['Revenue ($)'] - df_marketing['Spend ($)']) / df_marketing['Spend ($)']

print("DataFrame with new KPIs:")
display(df_marketing.head())
```

## Group by Channel

```python
# Group by channel
channel_perf = df_marketing.groupby('Channel').agg({
    'Spend ($)': 'sum',
    'Revenue ($)': 'sum',
    'Conversions': 'sum'
}).reset_index()

# Recalculate ROI for the aggregated channel performance
channel_perf['ROI (%)'] = ((channel_perf['Revenue ($)'] - channel_perf['Spend ($)']) / channel_perf['Spend ($)']) * 100

print("Channel Performance Metrics:")
display(channel_perf.sort_values(by='ROI (%)', ascending=False))
```

## Forecasting (Simple Moving Average)

For a simple forecast, we'll use a moving average on `Conversions`. First, we need to ensure the data is sorted by date.

```python
df_marketing_sorted = df_marketing.sort_values('Date')

# Calculate a 3-period moving average for Conversions
df_marketing_sorted['Forecast_Conversions_MA3'] = df_marketing_sorted['Conversions'].rolling(window=3).mean()

print("DataFrame with 3-period Moving Average Forecast for Conversions:")
display(df_marketing_sorted[['Date', 'Channel', 'Conversions', 'Forecast_Conversions_MA3']].head())
```
